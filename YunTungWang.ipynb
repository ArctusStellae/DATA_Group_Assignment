```r
url <- "https://data.ecan.govt.nz:443/data/98/Air/Air%20quality%20data%20for%20a%20monitored%20site%20(daily)/CSV?SiteId=10&StartDate=01%2F10%2F2013&EndDate=01%2F10%2F2023&zip=1"
response <- GET(url)
temp <- tempfile(fileext = ".zip")
writeBin(content(response, "raw"), temp)
unzip(temp, exdir = tempdir())
Timaru_Anzac_Square <- read_csv(file.path(tempdir(), "data.csv"))
```

```r
all_data <- bind_rows(Timaru_Anzac_Square, Rangiora, Ashburton, Geraldine, Kaiapoi)
data_list <- list(Timaru_Anzac_Square, Rangiora, Ashburton, Geraldine, Kaiapoi)
all_data <- bind_rows(data_list)
head(all_data)
```

```r
average_monthly_data <- all_data %>%
  group_by(month = month(DateTime)) %>% 
  summarise(`PM2.5 (ug/m3)` = mean(`PM2.5 (ug/m3)`, na.rm = TRUE), 
            `PM10 (ug/m3)` = mean(`PM10 (ug/m3)`, na.rm = TRUE),
            `Wind speed (m/s)` = mean(`Wind speed (m/s)`, na.rm = TRUE),
            `Temperature 2m (DegC)` = mean(`Temperature 2m (DegC)`, na.rm = TRUE),
            .groups = "drop")

long_data <- average_monthly_data %>% 
  gather(key = "Variable", value = "Value", -month)

pollutant_data <- long_data %>% filter(Variable %in% c("PM2.5 (ug/m3)", "PM10 (ug/m3)"))

wind_temp_data <- long_data %>% filter(Variable %in% c("Wind speed (m/s)", "Temperature 2m (DegC)"))

ggplot() +
  geom_col(data = pollutant_data, aes(x = factor(month), y = Value, fill = Variable), position = "dodge") +
  geom_line(data = wind_temp_data, aes(x = factor(month), y = Value, color = Variable, group = Variable), size = 1.5) +
  scale_fill_manual(values = c("PM2.5 (ug/m3)" = "lightblue", "PM10 (ug/m3)" = "darkblue")) +
  scale_color_manual(values = c("Wind speed (m/s)" = "orange", "Temperature 2m (DegC)" = "red")) +
  labs(title = "Monthly Average of Pollutants in South Island Over 10 Years",
       x = "Month", y = "Pollutants") +
  theme_minimal()
```

```r
long_data <- all_data %>% 
  pivot_longer(
    cols = c(`PM2.5 (ug/m3)`, `PM10 (ug/m3)`),
    names_to = "Pollutant",
    values_to = "Value"
  )
p <- ggplot(long_data, aes(x = StationName, fill = Pollutant, y = Value)) +
  geom_bar(stat = "identity", position = position_dodge()) + 
  labs(title = "Total Count Of PM2.5 and PM10 Levels Across Different Regions In 10 Years", 
       x = "Region", 
       y = "Total Count (ug/m3) In 10 Years") +
  theme_minimal(base_size = 10) +
  theme(
    plot.title = element_text(hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    panel.grid.major = element_line(colour = "white"),
    panel.grid.minor = element_line(colour = "white"),
    panel.background = element_rect(fill = "grey90"),
    axis.line = element_line(colour = "white")
  ) +
  scale_fill_brewer(palette = "Set3")

print(p)

```

```r
library(RColorBrewer)

color_palette <- brewer.pal(n = length(unique(all_data$StationName)), name = "Set3")

p <- ggplot(all_data, aes(x = DateTime, y = `PM10 (ug/m3)`, color = StationName)) +
  geom_smooth(se = FALSE, size = 1.2) + 
  labs(title = "Trend of PM10 Levels Over Time by City", 
       x = "Date", 
       y = "PM10 (ug/m3)") +
  theme_minimal(base_size = 14) + 
  theme(
    plot.title = element_text(face = "bold", size = 12, hjust = 0.5),
    axis.title = element_text(size = 10, face = "bold"),
    axis.text = element_text(size = 8),
    legend.title = element_text(face = "bold", size = 10),
    legend.text = element_text(size = 8),
    legend.position = "bottom",
    legend.box = "horizontal",
    panel.grid.major = element_line(colour = "grey85"),
    panel.grid.minor = element_line(colour = "grey95"),
    axis.line = element_line(colour = "black")
  ) + 
  scale_color_manual(values = color_palette)

print(p)
```
