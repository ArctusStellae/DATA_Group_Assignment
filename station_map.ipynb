```python

# Read the CSV file into a regular Pandas DataFrame
data = pd.read_csv('station_coordinates.csv')
```

```python
data.head()
```

```python
# Convert the DataFrame to a GeoDataFrame
geometry = [Point(xy) for xy in zip(data.site_long, data.site_lat)]
geo_df = gpd.GeoDataFrame(data, geometry=geometry)
```

```python
geo_df
```

```python
# Create an interactive map
Map = geemap.Map()

# Set a basemap: Options include 'ROADMAP', 'SATELLITE', 'TERRAIN', 'HYBRID'
Map.add_basemap('ROADMAP')

# Convert the GeoDataFrame to a FeatureCollection
features = []
for _, row in geo_df.iterrows():
    feature = ee.Feature(ee.Geometry.Point([row['site_long'], row['site_lat']]))
    features.append(feature)
fc = ee.FeatureCollection(features)

# Create an interactive map
Map = geemap.Map()

# Add the FeatureCollection to the map
Map.addLayer(fc, {}, 'Points from GeoDataFrame')

# Display the map
Map
```
